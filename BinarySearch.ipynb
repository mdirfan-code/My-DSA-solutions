### Binary search to find x in sorted array

```python
class Solution:
    def search(self, nums, target):
        
        l = 0
        r = len(nums)-1
        
        def bs(l,r,t):
            
            if(l>r):
                return -1
            
            m = l+(r-l)//2
            
            if(nums[m] == t):
                return m
            elif(nums[m] < t):
                return bs(m+1,r,t)
            else:
                return bs(l,m-1,t)
        return bs(l,r,target)
                
```

### Search Insert Position
Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

```python
class Solution:
    def searchInsert( nums, target):
        l = 0
        r = len(nums)-1
        ans = [r+1]
        def bs(l,r,t):
            
            if(l>r):
                return ans[0]
            
            m = l+(r-l)//2
            
            if(nums[m] == t):
                return m
            elif(nums[m] < t):
                return bs(m+1,r,t)
            else:
                ans[0] = m
                return bs(l,m-1,t)
        return bs(l,r,target)
```

### Floor/Ceil in Sorted Array


```python
def ceilingInSortedArray(n, x, arr):
    # Write your code here.
    arr.sort()
    l = 0
    r = len(arr)-1
    ans = [-1,-1]
    def bs(l,r,t):
        
        if(l>r):
            return -1
        
        m = l+(r-l)//2
        if arr[m] == t:
            ans[:] = [t,t]
            return 

        if(arr[m] > t):
            ans[1] = arr[m]
            return bs(l,m-1,t)
        else:
            ans[0] = arr[m]
            return bs(m+1,r,t)
       
    bs(l,r,x)
            
    return " ".join(list(map(str,ans)))
```

### Find the first and last occurance of a given number in a sorted array

```python
class Solution:
    def searchRange(nums, target):

        def bs(l,r,t):
            if l> r:
                return [-1,-1]
            m = l + (r - l) //2
            if nums[m] == t:
                i = m + 0
                j = m + 0
                while(i>0 and nums[i-1] == t ):
                    i-=1
                while(j<len(nums)-1 and nums[j+1] == t):
                    j+=1

                return [i,j]
            if nums[m] > t:
                return bs(l,m-1,t)
            else:
                return bs(m+1,r,t)
        return bs(0,len(nums)-1,target)
```

### Binary search with rotated array


```python
def search(nums, target) -> int:
    l = 0
    r = len(nums) - 1

    while l <= r:
        m = (l + r) // 2
        
        if nums[m] == target:
            return m

        if nums[l] <= nums[m]:
            if nums[l] <= target and target <= nums[m]:
                r = m - 1
            else:
                l = m + 1
        else:
            if nums[m] <= target and target <= nums[r]:
                l = m + 1
            else:
                r = m-1
    return -1
```

### Binary search with rotated array with duplicate values

```python
class Solution:
    def search(self, nums: List[int], target: int) -> bool:
        l = 0
        r = len(nums) - 1

        while l <= r:
            m = (l + r) // 2
            
            if nums[m] == target:
                return True
            #-------------------------
            if nums[l] == nums[m] and nums[m] == nums[r]:
                l += 1
                r -= 1
                continue
            #------------additional part-----------------------
            if nums[l] <= nums[m]:
                if nums[l] <= target and target <= nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                if nums[m] <= target and target <= nums[r]:
                    l = m + 1
                else:
                    r = m-1
        return False
```
